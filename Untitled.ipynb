```python
import matplotlib.pyplot as plt

# 讀取資料檔
with open("hw2_data.txt", "r") as file:
    lines = file.readlines()

# 統計每個單字出現次數
letter_counts = {}
for line in lines:
    word = line.strip().upper()
    if word:  # 如果不是空行
        letter_counts[word] = letter_counts.get(word, 0) + 1

# 印出統計結果（由多到少）
sorted_items = sorted(letter_counts.items(), key=lambda x: x[1], reverse=True)
for letter, count in sorted_items:
    print(f"{letter}: {count}")

# 畫圖
letters = [item[0] for item in sorted_items]
counts = [item[1] for item in sorted_items]

plt.bar(letters, counts)
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Letter Frequency Histogram")
plt.xticks(rotation=45)  # 避免字太擠
plt.tight_layout()
plt.savefig("histogram.png")
plt.show()
```
