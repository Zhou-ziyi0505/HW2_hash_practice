```python
import matplotlib.pyplot as plt

# ËÆÄÂèñË≥áÊñôÊ™î
with open("hw2_data.txt", "r") as file:
    lines = file.readlines()

# Áµ±Ë®àÂñÆÂ≠óÂá∫ÁèæÊ¨°Êï∏
word_counts = {}
for line in lines:
    word = line.strip().lower()  # Â∞èÂØ´ËôïÁêÜÈÅøÂÖçÈáçË§á
    if word:
        word_counts[word] = word_counts.get(word, 0) + 1

# Âª∫Á´ã‰∏çÈáçË§áÂñÆÂ≠óÈõÜÂêà
unique_words = set(word_counts.keys())

# È°ØÁ§∫Á∏ΩÂÖ±ÊúâÂπæÂÄã‰∏çÈáçË§áÁöÑÂñÆÂ≠ó
print(f"‚úÖ Á∏ΩÂÖ±Êúâ {len(unique_words)} ÂÄã‰∏çÈáçË§áÁöÑËã±ÊñáÂñÆÂ≠ó\n")

# Âç∞Âá∫ÊØèÂÄãÂñÆÂ≠óÂá∫ÁèæÊ¨°Êï∏ÔºàÁî±Â§öÂà∞Â∞ëÔºâ
print("üìä ÊØèÂÄãÂñÆÂ≠óÂá∫ÁèæÊ¨°Êï∏ÔºàÁî±Â§öÂà∞Â∞ëÔºâÔºö")
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_words:
    print(f"{word}: {count}")

# ‰ΩúÂúñÔºöÂá∫ÁèæÊ¨°Êï∏ÂæûÂ§öÂà∞Â∞ëÁöÑÁõ¥ÊñπÂúñ
words = [word for word, _ in sorted_words]
counts = [count for _, count in sorted_words]

plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency Histogram")
plt.tight_layout()
plt.savefig("histogram.png")
plt.show()
```
